```python
#bai 3
def happy_number(n, seen):
    ans = sum(int(i) ** 2 for i in str(n))
    if ans == 1:
        return True
    if ans in seen:
        return False
    seen.add(ans)
    return happy_number(ans, seen)


try:
    inp = input("Nhập vào một số nguyên dương a: ")
    a = int(input())
    if happy_number(a, set()):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Vui lòng nhập số nguyên hợp lệ!")
```

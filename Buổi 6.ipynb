```python
#bai 9
a=input().split()
b=input().split()
dic1={}
dic2={}
for i in range(0,len(a),2):
    dic1[a[i]]=int(a[i+1])
for j in range(0, len(b),2):
    dic2[b[j]]=int(b[j+1])
for k in dic2:
    if k in dic1:
        dic1[k]+=dic2[k]
    else:
        dic1["k"]=dic2[k]
print(dic1)
```

```python
# bai 17
n = int(input())
lst_chinh = []
lst_phu = []
for i in range(n):
    row = list(map(int,input().split()))
    for j in range(len(row)):
        if j == i:
            lst_chinh.append(row[j])
        if j == n - i - 1:
            lst_phu.append(row[j])
print(lst_chinh)
print(lst_phu)
```

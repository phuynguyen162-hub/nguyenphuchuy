```python
#W7A1
def binary_search(arr , left , right , target):
    l = left
    r = right
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            r = mid - 1
        else:
            l = mid + 1
    return -1

if __name__ == '__main__':
    arr = list(map(int,input().split()))
    target = int(input())
    print(binary_search(arr , 0 , len(arr) - 1 , target))
```

```python
#W7A2
def count_occurrences(arr , x):
    cnt = 0
    for i in arr:
        if i == x:
            cnt += 1
    return cnt

if __name__ == '__main__':
    arr = list(map(int,input().split()))
    x = int(input())
    print(count_occurrences(arr , x))
```

```python
#W7A3
def bubble_sort(arr):
    for i in range(0 , len(arr) - 1):
        for j in range(i + 1 , len(arr)):
            if arr[i] > arr[j]:
                arr[i] , arr[j] = arr[j] , arr[i]
    
    
if __name__ == '__main__':
    arr = list(map(int,input().split()))
    bubble_sort(arr)
    print(arr)
```

```python
#W7A5
def Dem(arr , x):
    dic = {}
    cnt = 0
    for i in arr:
        if x - i in dic:
            cnt += dic[x - i]
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1

    return cnt

if __name__ == '__main__':
    arr = list(map(int, input().split()))
    x = int(input())
    print(Dem(arr , x))
```

```python
#W7A9
st = [0] * int(1e5)

def get(id , l , r , u , v):
    if v < l or r < u:
        return 0
    if u <= l and r <= v:
        return st[id]
    mid = (l + r) // 2
    return get(id * 2 , l , mid , u , v) + get(id * 2 + 1 , mid + 1 , r , u , v)

def update(id , l , r , pos , val):
    if r < pos or pos < l:
        return
    if l == r:
        st[id] += val
        return
    mid = (l + r) // 2
    update(id * 2 , l , mid , pos , val)
    update(id * 2 + 1 , mid + 1 , r , pos , val)
    st[id] = st[id * 2] + st[id * 2 + 1]    

if __name__ == '__main__':
    arr = list(map(int , input().split()))
    res = [0] * len(arr)
    n = max(arr)
    for i in range(len(arr) - 1 , -1 , -1):
        ans = get(1 , 0 , n + 1 , 0 , arr[i] - 1)
        
        update(1 , 0 , n + 1 , arr[i] , 1)
        res[i] = ans
    print(res)

```

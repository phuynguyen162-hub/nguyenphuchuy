```python
#bai1
def Merge(a,b):
    c=a+b
    n = len(c)
    for i in range(len(c)):
        for j in range(n-i-1):
            if c[j]>c[j+1]:
                c[j], c[j+1]=c[j+1], c[j]
    return c
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*Merge(a,b))
```

```python
#bài 2
def count_less_equal(matrix, n, m, target):
    count = 0
    row = n - 1
    col = 0
    while row >= 0 and col < m:
        if matrix[row][col] <= target:
            count += (row + 1)
            col += 1
        else:
            row -= 1
    return count


def find_kth_smallest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])

    low = matrix[0][0]
    high = matrix[n - 1][m - 1]

    while low < high:
        mid = (low + high) // 2
        if count_less_equal(matrix, n, m, mid) < k:
            low = mid + 1
        else:
            high = mid

    return low

matrix = []
k = 0

while True:
    line = list(map(int, input().split()))
    if len(line) == 1:
        k = line[0]
        break
    matrix.append(line)

print(find_kth_smallest(matrix, k))
```

```python
#bai 4
def solve():
    q_input = input()
    if not q_input:
        return
    q = int(q_input)

    for _ in range(q):

        a = input()
        b = input()

        n = len(a)
        m = len(b)

        dp = [[False] * (m + 1) for _ in range(n + 1)]
        dp[0][0] = True

        for i in range(1, n + 1):
            if a[i - 1].islower():
                dp[i][0] = dp[i - 1][0]
            else:
                dp[i][0] = False

        for i in range(1, n + 1):
            for j in range(1, m + 1):
                char_a = a[i - 1]
                char_b = b[j - 1]

                if char_a.isupper():

                    if char_a == char_b:
                        dp[i][j] = dp[i - 1][j - 1]
                    else:
                        dp[i][j] = False
                else:
                    option_delete = dp[i - 1][j]
                    option_convert = False
                    if char_a.upper() == char_b:
                        option_convert = dp[i - 1][j - 1]

                    dp[i][j] = option_delete or option_convert

        if dp[n][m]:
            print("YES")
        else:
            print("NO")

solve()
```

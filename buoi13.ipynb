```python
#bai 1
def reverse(path: str) -> int:
    try:
        with open(path, 'r') as f:
            s = f.read().strip()
            if s.startswith('-'):
                reversed_num = int('-' + s[:0:-1])
            else:
                reversed_num = int(s[::-1])
            return reversed_num
    except FileNotFoundError:
        return 0
     
```

```python
#bai 5
def count(path: str, character: str) -> int:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            x = f.read()
        return x.count(character)
    except Exception:
        return 0
```

```python
#bai 11
def find_max_min(path: str):
    try:
        with open(path, 'r') as f:
            content = f.read().replace('\n', ' ').split()
            nums = [int(x) for x in content]
            if not nums:
                return
            print(f"{max(nums)} {min(nums)}")
    except Exception:
        print("Failed")
```

```python
#bai 12
def maximumProduct(path: str) -> int:
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            nums = list(map(int, lines[1].split()))
            nums.sort()
            return max(nums[-1] * nums[-2] * nums[-3], nums[0] * nums[1] * nums[-1])
    except:
        return 0
```

```python
#bai 13
def move(fileName: str) -> list:
    try:
        with open(fileName, 'r') as f:
            lines = f.readlines()
            nums = list(map(int, lines[1].split()))
        non_zero = [x for x in nums if x != 0]
        zeros = [0] * (len(nums) - len(non_zero))
        return non_zero + zeros
    except:
        return []
```

```python
#bai 14
def average_time(path: str) -> float:
    try:
        total_time = 0.0
        count = 0
        with open(path, 'r') as f:
            reader = csv.DictReader(f)
            reader.fieldnames = [name.strip() for name in reader.fieldnames]
            for row in reader:
                total_time += float(row['Time'])
                count += 1

        if count == 0:
          return 0.0
        return round(total_time / count, 2)
    except:
        return 0.0
```

```python
#bai 16
def calculate_geometry_perimeter(path: str):
    try:
        with open(path, 'r') as f:
            content = f.read().split('\n\n')

        for i in content:
            lines = i.strip().split('\n')
            if not lines:
              continue
            shape_type = lines[0].split(':')[1].strip()
            x = lines[1]

            if shape_type == 'SQUARE':
                side = float(x.split(':')[1])
                print(f"{side * 4:.2f}")

            elif shape_type == 'RECTANGLE':
                parts = x.split()
                width = float(parts[1])
                height = float(parts[3])
                print(f"{(width + height) * 2:.2f}")

            elif shape_type == 'CIRCLE':
                radius = float(x.split(':')[1])
                print(f"{2 * math.pi * radius:.2f}")

    except Exception as e:
        pass
     
```

```python
#bai 17
def product(path: str) -> list[int]:
    try:
        with open(path, 'r') as f:
            content = f.read().split()
            if len(content) > int(content[0]) + 1:
                 nums = [int(x) for x in content[1:]]
            else:
                 nums = [int(x) for x in content if x.isdigit()]
                 nums = [int(x) for x in content[1:]]

        n = len(nums)
        res = [1] * n

        left = 1
        for i in range(n):
            res[i] = left
            left *= nums[i]

        right = 1
        for i in range(n - 1, -1, -1):
            res[i] *= right
            right *= nums[i]

        return res
    except:
        return []
```

```python
#bai 18
def cal(path: str) -> float:
    try:
        with open(path, 'r') as f:
            lines = f.readlines()

        src_parts = lines[0].split()
        src_x, src_y = float(src_parts[2]), float(src_parts[4])

        des_parts = lines[1].split()
        des_x, des_y = float(des_parts[2]), float(des_parts[4])

        vel = float(lines[2].split()[1])

        distance = math.sqrt((des_x - src_x)**2 + (des_y - src_y)**2)
        return round(distance / vel, 2)
    except:
        return 0.0
```

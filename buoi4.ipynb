```python
#bài 8
def reverse(n):
    a=0
    lst=[]
    reverse=0
    while n>=1:
        a=n%10
        n=n//10
        lst.append(a)
    for i in lst:
        reverse=reverse*10+i
    return reverse
n=int(input())
x=0
while n!= reverse(n):
    n+=reverse(n)
    x+=1
print(x, n)
```

```python
#bài 26
def fibo(n):
    if n<=0:
        return []
    else:
        lst=[1,1]
        while max(lst)<=n:
            a=lst[len(lst)-2]+lst[len(lst)-1]
            lst.append(a)
        return lst
n=int(input())
print(fibo(n)[len(fibo(n))-2])
    
    
```

```python
#W5A1
def Max_Of_True(a,b):
    if a>b:
        return a
    else:
        return b
a,b=map(int,input().split())
print(Max_Of_True(a,b))
```

```python
#W5A2
def swap(a,b):
    return b,a
a,b=map(int,input().split())
print(swap(a,b))
```

```python
#W5A3
def prime(n):
    if n<2:
        return False
    else:
        for i in range(2,int((n**0.5)+1)):
            if n%i==0:
                return False
        return True
n=int(input())
print(prime(n))
```

```python
#W5A4
def hoanhao(n):
    tong=0
    for i in range(1,n+1):
        if n%i==0:
            tong+=i
        if tong==n:
            return True
    return False
n=int(input())
print(hoanhao(n))
```

```python
#W5A5
def vitri(ds,k):
    for i in range(0,len(ds)):
        if ds[i]==k:
            return i+1
    return -1
ds=list(map(int,input().split()))
k=int(input())
print(vitri(ds,k))
```

```python
#W5A6
def factorial(n):
    kq=1
    for i in range(1,n+1):
        kq*=i
    return kq
n=int(input())
print(factorial(n))
```

```python
#W5A7
def maytinh(n):
    a=""
    b=""
    for i in range(len(n)):
        if n[i]=="+":
            a=n[:i]
            b=n[i:]
            return float(a)+float(b)
        elif n[i]=="-":
            a=n[:i]
            b=n[i:]
            return float(a)-float(b)
        elif n[i]=="*":
            a=n[:i]
            b=n[i+1:]
            return ifloat(a)*ifloat(b)
        elif n[i]=="/":
            a=n[:i]
            b=n[i+1:]
            return round(float(a)/float(b),2)
        
n=input()
print(maytinh(n))


    
```

```python
#W5A8
def hicut(n):
    lst=[]
    while n>0:
        lst.append(n%2)
        n=n//2
    lst.reverse()
    return lst
a,b=map(int,input().split())
x=hicut(a)
y=hicut(b)
while len(x)>len(y):
    y.insert(0,0)
while len(y)>len(x):
    x.insert(0,0)
t=0
for i in range(len(x)):
    if x[i]!=y[i]:
        t+=1
print(t)
```

```python
#bai
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def dientich(self):
        return 3.14*(self.r**2)*2+3.14*2*self.r*self.h
    def thetich(self):
        return 3.14*self.r**2*self.h
n = input().split(',')
p = Cylinder(int(n[0]), int(n[1]))
print("Diện tích hình trụ:", p.dientich())
print("Thể tích hình trụ:", p.thetich())

```

```python
#bai2
class Date:
    def __init__(self,d,m,y):
        self.d=d
        self.m=m
        self.y=y
    def nextday(self):
        if self.y%4==0 and self.y%100!=0 or self.y%400==0:
            ngaythang=(31,29,31,30,31,30,31,31,30,31,30,31)
        else:
            ngaythang=(31,28,31,30,31,30,31,31,30,31,30,31)
        self.d+=1
        if self.d > ngaythang[self.m-1]:
            self.d = 1
            self.m += 1
        if self.m > 12:
            self.y +=1
            self.m = 1
        return f"{self.d, self.m, self.y}"
n=input().split('/')
p = Date(int(n[0]), int(n[1]), int(n[2]))
print(p.nextday())
  
```

```python
class ShoppingCart:
  def __init__(self):
    self.items = []
  def add(self,name,price):
    self.items.append((name,price))
  def remove_item(self,name,price):
    try:
      self.items.remove((name, price))
      print("Đã xóa sản phẩm")
    except ValueError:
      print("Không tìm thấy sản phẩm")
  def check(self):
    if not self.items:
      return 'Giỏ hàng rỗng'
    else:
      return 'Giỏ hàng không rỗng'
  def sum_price(self):
    return sum(y for x,y in self.items)
  def show_all(self):
    if not self.items:
      print('Giỏ hàng rỗng')
    else:
      for name,price in self.items:
        print(f'{name} - {price}')
  def remove_all(self):
    self.items = []
    print('Đã xóa thành công')
  def menu(self):
    print('1:Thêm vào giỏ hàng')
    print('2:Xóa khỏi giỏ hàng')
    print('3:Kiểm tra giỏ hàng')
    print('4:Tổng giá trị')
    print('5:Hiển thị toàn bộ giỏ hàng')
    print('6:Xóa toàn bộ giỏ hàng')
    n=input('Chọn mục : ')
    while 1:
      if n == '1':
        self.add(input('Nhập tên sản phẩm '),int(input('Nhập giá sản phẩm ')))
      elif n == '2':
        self.remove_item(input('Nhập tên sản phẩm '),int(input('Nhập giá sản phẩm ')))
      elif n == '3':
        print(self.check())
      elif n == '4':
        print(self.sum_price())
      elif n == '5':
        self.show_all()
      elif n == '6':
        self.remove_all()
      elif n == 'yes':
        break
      n=input('Chọn mục : ')
s=ShoppingCart()
s.menu()
```

```python
#bai 6
class Rectangle:
    def __init__(self,w,h,k):
        self.w=w
        self.h=h
        self.k=k
    def area(self):
        return self.w*self.h*(self.k**2)
    def perimeter(self):
        return 2*(self.w+self.h)*self.k

n = input().split(',')
p = Rectangle(int(n[0]), int(n[1]), int(n[2]))
print("Area:", p.area())
print("Perimeter:", p.perimeter())
```

```python
#bai 7
def UCLN(num, den):
    if den == 0:
        return num
    return UCLN(den,num%den)
    
class Fraction:
    def __init__(self , num , den):
        self.num = int(num)
        self.den = int(den) 
    
    def __add__(self , other):
        tu_so = self.num * other.den + self.den * other.num
        mau_so = self.den * other.den
        return tu_so , mau_so

    def __sub__(self , other):
        tu_so = self.num * other.den - self.den * other.num
        mau_so = self.den * other.den
        return tu_so , mau_so

    def __mul__(self , other):
        tu_so = self.num * other.num
        mau_so = self.den * other.den
        return tu_so , mau_so

    def __div__(self , other):
        tu_so = self.num * other.den
        mau_so = self.den * other.num
        return tu_so , mau_so
    
    def __str__(self):
        return (f"{self.num}/{self.den}")
    
s = input().split()
A = Fraction(s[0] , s[1])
B = Fraction(s[3] , s[4])
if s[2] == '+':
    a , b = A + B
elif s[2] == '-':
    a , b = A - B
elif s[2] == '*':
    a , b = A * B
elif s[2] == '/':
    a , b = A / B
 
ucln = UCLN(a , b)
a /= ucln
b /= ucln

res = Fraction(a , b)
print(res)
```

```python
#bai 8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        if amount > self.balance:
            pass
        else:
            self.balance -= amount
    def __str__(self):
        return f"{self.balance:.2f}"
owner, balance = input().split()
account = BankAccount(owner, float(balance))
n = int(input())
for _ in range(n):
    action, amount = input().split()
    amount = float(amount)
    if action == "DEPOSIT":
        account.deposit(amount)
    elif action == "WITHDRAW":
        account.withdraw(amount)
print(account.__str__())
```

```python
#bai 10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.books = []
    @classmethod
    def from_string(cls, book_str):
        lst = book_str.split(';')
        title = lst[0]
        author = lst[1]
        year = int(lst[2])
        return Book(title, author, year)
    def add_book(self, info):
        book = Library.from_string(info)
        self.books.append(book)
    def count_by_author(self, author_name):
        cnt = 0
        for book in self.books:
            if book.author == author_name:
                cnt += 1
        return cnt
    def find_by_year(self, year):
        cnt = 0
        for book in self.books:
            if book.year == year:
                cnt += 1
        return cnt
n = int(input())
library = Library()
for _ in range(n):
    info = input().split()
    if info[0] == "ADD":
        library.add_book(info[1])
    elif info[0] == "COUNT":
        author_name = info[1]
        print(library.count_by_author(author_name))
    elif info[0] == "COUNT_YEAR":
        year = int(info[1])
        print(library.find_by_year(year))
```

```python
#bai 11
import math

class Degrees:
    def __init__(self , A , B , C):
        self.A = A
        self.B = B
        self.C = C
        self.pt = []

    def vector_pt(self):
        A = self.A
        B = self.B
        C = self.C
        AB = [B[0] - A[0] , B[1] - A[1] , B[2] - A[2]]
        AC = [C[0] - A[0] , C[1] - A[1] , C[2] - A[2]]
        self.pt = [AB[1] * AC[2] - AC[1] * AB[2] , AB[2] * AC[0] - AC[2] * AB[0] , AB[0] * AC[1] - AC[0] * AC[1]]

    def tinh_goc(self , other):
        pt1 = self.pt
        pt2 = other.pt
        x = (abs(pt1[0] * pt2[0] + pt1[1] * pt2[1] + pt1[2] * pt2[2])) / ((pt1[0] ** 2 + pt1[1] ** 2 + pt1[2] ** 2) ** 0.5 * (pt2[0] ** 2 + pt2[1] ** 2 + pt2[2] ** 2) ** 0.5)
        return x
    #def tinh_goc(self , other)

if __name__ == '__main__':
    A = list(map(float,input().split()))
    B = list(map(float,input().split()))  
    C = list(map(float,input().split()))
    D = list(map(float,input().split()))
    mp1 = Degrees(A , B , C)
    mp2 = Degrees(B , C , D) 
    mp1.vector_pt()
    mp2.vector_pt()
    x = mp1.tinh_goc(mp2)
    x = math.acos(x)
    print(f'{math.degrees(x):.2f}')
    
```

```python
#bai 13
import math
from math import sqrt
class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    @property
    def check(self):
        if self.edge1 + self.edge2 > self.edge3 and self.edge2 + self.edge3 > self.edge1 and self.edge1 + self.edge3 > self.edge2:
            return True
        else:
            return False
    def get_area(self):
        p = (self.edge1 + self.edge2 + self.edge3)/2
        s = sqrt(p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))
        return f"{s:.2f}"
tri = Triangle(*map(int,input().split()))
if tri.check:
    print(tri.get_area())
else:
    print("Invalid")
```
